```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
X = np.array([[x] for x in range(1, 101)])
```

```python
X.shape
```

```python
X
```

```python
y = np.array([x*3 + np.random.rand()*10 for x in X])
```

```python
plt.figure(figsize=(16,7))
plt.scatter(X, y)
```

```python
init_guess = np.array([[2]])
```

```python
init_guess
```

```python
init_guess.shape
```

```python
X.shape, init_guess.shape
```

```python
ypred = np.dot(X, init_guess)
```

```python
ypred.shape
```

```python
plt.figure(figsize=(16,8))

plt.scatter(X, y)
plt.scatter(X, ypred)
```

```python
init_guess = np.array([[4]])
ypred = np.dot(X, init_guess)

plt.figure(figsize=(16,8))

plt.scatter(X, y)
plt.scatter(X, ypred)
```

```python
weights = np.linspace(2,4,100)
```

```python
weights
```

```python
for w in weights:
    ypred = np.dot(X, np.array([[w]]))
    plt.scatter(X, y)
    plt.scatter(X, ypred)
    
    plt.show()
```
