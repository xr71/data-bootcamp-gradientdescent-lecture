```python
import matplotlib.pyplot as plt

%matplotlib inline
```

$$
X - A^2 \approx 0
$$

Where A is my guess, iteratively...

```python
X = 121
init_guess = 1
```

```python
error = X - (init_guess * init_guess)
```

```python
lr = 0.001
```

```python
grad = error * lr
```

```python
print(error, grad)
```

```python
init_guess += grad
```

```python
print(f"After one manual training loop, my theta is now {init_guess}")
```

```python
def approx_sqrt_1(X, initguess=1, numiter=2, lr=0.001):
    
    for i in range(numiter):
        ypred = initguess * initguess
        err = X - ypred
        grad = err
        step = grad * lr
        
        print(round(ypred, 2), round(err, 4), round(step, 4), round(initguess, 3))
        
        initguess += step
```

```python
approx_sqrt_1(121, numiter=5)
```

```python
approx_sqrt_1(121, numiter=20, lr=0.005)
```

```python
def approx_sqrt_2(X, initguess=1, numiter=2, lr=0.001):
    
    for i in range(numiter):
        ypred = initguess * initguess
        err = X - ypred
        grad = err
        step = grad * lr
        
        initguess += step
        
        yield initguess
```

```python
for g in approx_sqrt_2(121):
    print(g)
```

```python
eStepGuess = list(approx_sqrt_2(121, numiter=20))

print(eStepGuess[-1])

plt.figure(figsize=(16,7))
plt.plot(eStepGuess)
```

```python
def approx_sqrt_3(X, initguess=1, numiter=2, lr=0.001):
    
    for i in range(numiter):
        ypred = initguess * initguess
        err = X - ypred
        grad = err
        step = grad * lr
        
        initguess += step
        
        yield initguess, err
```

```python
approx_sqrt_3(121)
```

```python
list(approx_sqrt_3(121))
```

```python
guessWithError = list(approx_sqrt_3(121, numiter=100, lr=0.005))

guesses = [x[0] for x in guessWithError]
errors = [x[1] for x in guessWithError]

plt.plot(errors)

plt.plot(guesses)
plt.show()
```
